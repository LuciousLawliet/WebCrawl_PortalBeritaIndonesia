```python
from requests_html import HTMLSession
import numpy as np
import requests
from requests import get
from bs4 import BeautifulSoup
import time
import random

web_name = input('Masukkan nama web: ')
keyword = input('Masukkan kata yang ingin dicari: ')
pg = input('Berapa halaman yang ingin dicari: ')
pg = int(pg) + 1

newslist = []
pages = np.arange(1,int(pg),1)
web_name = web_name.lower()

#cnbc
if web_name == 'cnbc' or web_name == 'cnbc indonesia' or web_name == 'cnbcindonesia':
  session = HTMLSession()
  for page in pages:
    r = session.get('https://www.cnbcindonesia.com/search?query=' + str(keyword) + '&p=' + str(page) + '&kanal=&tipe=&date=')
    articles = r.html.find('article')
    for item in articles:
      try:
        newsitem = item.find('a', first=True)
        title = newsitem.text
        link = newsitem.absolute_links
        newsarticle = {
            'title': title,
            'link': link
        }
        newslist.append(newsarticle)
      except:
        pass
  print(len(newslist))
  for c in newslist:
    print(c)
#antara news
elif web_name == 'antaranews' or web_name == 'antara news':
  for page in pages:
    time.sleep(random.randint(1,10))
    url = 'https://www.antaranews.com/search/' + str(keyword) + '/' + str(page)
    results = requests.get(url)
    soup = BeautifulSoup(results.text, 'html.parser')
    news = soup.find_all('article', class_='simple-post simple-big clearfix')
    for container in news:
      headline = container.header.h3.a.text
      link = container.header.h3.a['href']
      newsarticle = {
          'title': headline,
          'link': link
      }
      newslist.append(newsarticle)
  print(len(newslist))
  for i in newslist:
    print(i)
#bisnis.com
elif web_name == 'bisnis.com' or web_name == 'bisnis' or web_name == 'ekonomi bisnis':
  for page in pages:
    time.sleep(random.randint(1,10))
    url = 'https://search.bisnis.com/?q='+ str(keyword) +'&per_page=' + str(page)
    results = requests.get(url)
    soup = BeautifulSoup(results.text, 'html.parser')
    news = soup.find_all('div', class_='col-sm-8')
    for container in news:
        headline = container.h2.a.text
        link = container.h2.a['href']        
        newsarticle = {
            'title': headline,
            'link': link
        }
        newslist.append(newsarticle)
  print(len(newslist))
  for i in newslist:
    print(i)
#investor.id
elif web_name == 'investorid' or web_name == 'investor id' or web_name == 'investor.id':
  for page in pages:
    time.sleep(random.randint(1,10))
    url = 'https://investor.id/search/?web_name=' + str(keyword) + '&page=' + str(page)
    results = requests.get(url)
    soup = BeautifulSoup(results.text, 'html.parser')
    news = soup.find_all('div', class_='listview-content pb10 mb20')
    for container in news:
        headline = container.div.p.a.text
        link = container.div.p.a['href']        
        newsarticle = {
            'title': headline,
            'link': link
        }
        newslist.append(newsarticle)
  print(len(newslist))
  for i in newslist:
    print(i)
#medcom.id
elif web_name == 'medcom' or web_name == 'medcom.id':
  pages = np.arange(1,int(pg),20)
  for page in pages:
    time.sleep(random.randint(1,10))
    url = 'https://m.medcom.id/search?q=' + str(keyword) + '&per_page=' + str(page)
    results = requests.get(url)
    soup = BeautifulSoup(results.text, 'html.parser')
    news = soup.find_all('div', class_='block')
    for container in news:
        headline = container.a['title']
        link = container.a['href']     
        newsarticle = {
            'title': headline,
            'link': link
        }
        newslist.append(newsarticle)
  print(len(newslist))
  for i in newslist:
    print(i)
#minews.id
elif web_name == 'minews' or web_name == 'minews.id' or web_name == 'mi news':
  for page in pages:
    time.sleep(random.randint(1,10))
    url = 'https://www.minews.id/page/' + str(page) + '?s=' + str(keyword)
    results = requests.get(url)
    soup = BeautifulSoup(results.text, 'html.parser')
    news = soup.find_all('h3', class_='entry-title td-module-title')
    for container in news:
        headline = container.a['title']
        link = container.a['href']     
        newsarticle = {
            'title': headline,
            'link': link
        }
        newslist.append(newsarticle)
  print(len(newslist))
  for i in newslist:
    print(i)
#harian haluan
elif web_name == 'harianhaluan' or web_name == 'harian haluan' or web_name == 'haluan.com' or web_name == 'haluan':
  for page in pages:
    time.sleep(random.randint(1,10))
    url = 'https://www.harianhaluan.com/search?q=' + str(keyword) + '&page=' + str(page)
    results = requests.get(url)
    soup = BeautifulSoup(results.text, 'html.parser')
    news = soup.find_all('h2', class_='latest__title')
    for container in news:
        headline = container.a.text
        link = container.a['href']     
        newsarticle = {
            'title': headline,
            'link': link
        }
        newslist.append(newsarticle)
  print(len(newslist))
  for i in newslist:
    print(i)
#detiknews
elif web_name == 'detik' or web_name == 'detiknews':
  for page in pages:
    time.sleep(random.randint(1,10))
    url = 'https://www.detik.com/search/searchall?query=' + str(keyword) + '&siteid=2&sortby=time&page=' + str(page)
    results = requests.get(url)
    soup = BeautifulSoup(results.text, 'html.parser')
    news = soup.find_all('article')
    page_article = np.arange(1,4,1)
    for container in news:
        isi = []
        headline = container.a.h2.text
        link = container.a['href']
        for halaman in page_article:    
          results = requests.get(link + '/' + str(halaman))
          soup = BeautifulSoup(results.text)
          news = soup.find_all('p')
          for berita in news:
            p = berita.text
            isi.append(p)
          paragraph = ' '.join(isi)
          newsarticle = {
              'title': headline,
              'link': link,
              'paragraph': paragraph
          }
        newslist.append(newsarticle)
  print(len(newslist))
  for i in newslist:
    print(i)
#sindonews
elif web_name == 'sindonews' or web_name == 'sindo' or web_name == 'sindonews.com':
  pages = np.arange(0,int(pg*12),20)
  for page in pages:
    time.sleep(random.randint(1,10))
    url = 'https://search.sindonews.com/go/' + str(page) + '?type=artikel&q=' + str(keyword)
    results = requests.get(url)
    soup = BeautifulSoup(results.text, 'html.parser')
    news = soup.find_all('div', class_='news-title ekbis-title')
    page_article = np.arange(1,4,1)
    for container in news:
      isi = []
      headline = container.a.text
      link = container.a['href']
      for halaman in page_article:
          results = requests.get(link + '/' + str(halaman))
          soup = BeautifulSoup(results.text)
          news = soup.find_all('div', class_='vidy-embed')
          for berita in news:
            p = berita.text
            isi.append(p)
          paragraph = ''.join(isi)
          newsarticle = {
            'title': headline,
            'link': link,
            'paragraph': paragraph
        }
      newslist.append(newsarticle)
  print(len(newslist))
  for i in newslist:
    print(i)
```
